```python
import torch
import pandas as pd
from transformers import MarianMTModel, MarianTokenizer
from tqdm import tqdm

# Load pre-trained MarianMT model and tokenizer
MODEL_NAME = "Helsinki-NLP/opus-mt-en-it"

# Check if GPU is available for faster processing
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load tokenizer and model
tokenizer = MarianTokenizer.from_pretrained(MODEL_NAME)
model = MarianMTModel.from_pretrained(MODEL_NAME).to(device)


def translate_titles(titles, tokenizer, model, batch_size=5):
    """
    Translate a list of English products to Italian.

    Args:

    titles (list): List of English product titles.
    tokenizer: Pre-trained MarianTokenizer.
    model: Pre-trained MarianMTModel.
    batch_size (int): Number of titles processed at once.

    Returns:
    list: List of translated Italian titles.
    """

    translated_titles = []

    for i in tqdm(range(0, len(titles), batch_size), desc='Translating Titles'):
        batch = titles[i:i + batch_size]
        tokenized_batch = tokenizer(batch, return_tensors= "pt", padding=True, truncation=True).to(device)
        translated_tokens = model.generate(**tokenized_batch)
        translated_batch = [tokenizer.decode(tokens, skip_special_tokens=True) for tokens in translated_tokens]
        translated_titles.extend(translated_batch)

    return translated_titles




# Translate English titles to Italian
df['title_italian'] = translate_titles(df['title'].tolist(), tokenizer, model)

```

```python
df['productDescription_italian'] = translate_titles(df['product_description'].tolist(), tokenizer, model)
```

```python
import pandas as pd
from langdetect import detect
from nltk.tokenize import sent_tokenize

# Sample DataFrame with mixed titles
data = pd.read_csv(r"C:\Users\Charran\Documents\GitHub\EE6405-Weekly-Activities\Target.csv")
df = pd.DataFrame(data)
df = df[['title'],['product description']]
from langdetect import detect_langs


import re

# Apply the advanced function to the "product_name" column
def split_mixed_languages(text):
    if pd.isna(text):  # Check if the text is NaN
        return {
            'english_text': '',
            'spanish_text': ''
        }
    
    
    english = []
    spanish = []
    #words = str(text).split()  # Convert to string before splitting
    words = re.findall(r'\b[a-zA-Z]+\b', text)  # Extract only alphabetic words
    for word in words:
        try:
            # Detect language probabilities for each word
            lang_probs = detect_langs(word)
            primary_lang = lang_probs[0].lang
            
            if primary_lang == 'en':
                english.append(word)
            elif primary_lang == 'es':
                spanish.append(word)
        except:
            continue
    
    return {
        'english_text': ' '.join(english),
        'spanish_text': ' '.join(spanish)
    }

# Apply the advanced function to the "product_name" column
df = df.join(df['product_name'].apply(lambda x: pd.Series(split_mixed_languages(x))))

# Display the result
print(df[['english_text', 'spanish_text']])
```
